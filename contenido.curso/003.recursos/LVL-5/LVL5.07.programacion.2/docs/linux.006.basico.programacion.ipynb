**QUE ES UN ALGORITMO?**

Un algoritmo es una serie de instrucciones secuenciales ―es decir, que van uno después del otro



Programación es el proceso de tomar un algoritmo y codificarlo en una notación, un lenguaje de programación, de modo que pueda ser ejecutado por una computadora. Aunque existen muchos lenguajes de programación y muchos tipos diferentes de computadoras, el primer paso es la necesidad de tener una solución. Sin un algoritmo no puede haber un programa.


**QUE ES BASH SCRIPTING?**


Bash scripting es la fusión perfecta entre programación y el intérprete de comandos más común de Linux.

El uso de bash scripting permite:

1.   Automatizar acciones repetitivas, gracias al uso de expresiones matemáticas, como condiciones, bucles puertas lógicas, etc.

2.   Como consecuencia de lo anterior, nos permite mejorar la experiencia del usuario.

3.   También ofrece las herramientas necesarias a un administrador para que su sistema operativo sea más automático, más ágil y más capaz de procesar datos.

**CONDICIONELES**

Un condicional en la programación es una sentencia o grupo de sentencias que puede ejecutarse o no en función del valor de una condición. 


1.   En bash los condicionales numericos y de texto se hacen diferente

```
  # validaciones para textos

  cadena == cadena | --> recomiendo esta
  cadena = cadena

  # validaciones para numeros
  x -lt y			x menor que y
  x -le y			x menor o igual que y
  x -eq y			x igual que y
  x -ge y			x mayor o igual que y
  x -gt y			x mayor que y
  x -ne y			x no igual que y
```

2.   En los otros lengajes las validaciones se realizan unicamente con `==`

3.   si se tienen varias conficiones para el mismo segmento, cuando encuentra una ignora las demas

3.  se pueden tener cuantas condiciones rerequiera

Nota: bash es como un abuelo, muy permisivo en algunas cosas, pero muy obstinado en otras


````python
%%bash
# validaciones de strings | fi solo

if [ "aa" == "b" ]; then
  echo "entro......!!"
fi
````

````python
%%bash
# validaciones de strings | if - else

if [ "aa" == "b" ]; then
  echo "[CUMLPE] entro......!!"
else
  echo "[NO CUMPLE] ....."
fi
````

**BUCLES**

Un bucle o ciclo, en programación, es una secuencia de instrucciones de código que se ejecuta repetidas veces, hasta que la condición asignada a dicho bucle deja de cumplirse. Los tres bucles más utilizados en programación son el bucle while, el bucle for y el bucle do-while.

**FOR**

Un for en programación se usa cuando queremos repetir un conjunto de instrucciones un número finito(numeros de elementos que tenga la lista) de veces.


````python
%%bash
numeros=(1 2 3 4 5)
# ${numeros[@]} | 

for numero in ${numeros[@]}; do
  # a cada elemento de la lista le aplicara la logica que tenga el for
  echo "el numero es : $numero"
  echo "---------------------"
done
````

````python
%%bash
ls /tmp
````

````python
%%bash
for archivo in $(ls /tmp); do
  # a cada elemento de la lista le aplicara la logica que tenga el for
  echo "el archivo es : $archivo"
  echo "---------------------"
done
````

````python
%%bash
echo "primera linea" > log.log
echo "otra linea" >> log.log
echo "linea final" >> log.log

for linea_log in $(cat log.log); do
  # a cada elemento de la lista le aplicara la logica que tenga el for
  echo "la linea es : $linea_log"
  echo "---------------------"
done
````

**WHILE**

El bucle while o bucle mientras es un ciclo repetitivo basado en los resultados de una expresión lógica; se encuentra en la mayoría de los lenguajes de programación estructurados. El propósito es repetir un bloque de código mientras una condición se mantenga verdadera.1​


1.   El while es infinito(se va a ejecutar hasta que se cumpla una condición)
1.   En algunos casos es finito(lectura de archivo)


````python
%%bash
# suma
suma=1
suma=$((suma+1))  # ok
let suma=$suma+1  # ok
echo "el valor de la suma es : $suma"
````

````python
%%bash
# bucle infinito
numero=0
while [ $numero -le 10 ]; do
  echo "[ITERO]... el numero es $numero" # direccionamiento
done
````

````python
%%bash
# bucle infinito
numero=0
while [ $numero -le 10 ]; do
  echo "[ITERO]... el numero es $numero" # direccionamiento
  let numero=$numero+1
done
````

````python
%%bash
# > redireccionamiento a un archivo
# < redireccionamiento de un archivo a un comando
while read linea; do
  echo "contenido de la linea : --> $linea"
done < log.log
````
